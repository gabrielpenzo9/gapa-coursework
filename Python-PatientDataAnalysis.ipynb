```python
def calculate_cardio_risk(row):
    if row['Diastolic BP'] < 80 and row['Systolic BP'] < 120 and row['Age'] < 50:
        return 'Low'
    elif (80 <= row['Diastolic BP'] < 90 or 
          120 <= row['Systolic BP'] < 140 or 
          50 <= row['Age'] <= 60):
        return 'Moderate'
    else:
        return 'High'

final_merged['Cardio Risk'] = final_merged.apply(calculate_cardio_risk, axis=1)

print(final_merged[['Patient ID', 'Age', 'Systolic BP', 'Diastolic BP', 'Cardio Risk']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

group1 = final_merged[final_merged['Smoking Status'] == 'Current Smoker']['Cholesterol']
group2 = final_merged[final_merged['Smoking Status'] == 'Never']['Cholesterol']

t_stat, p_val = stats.ttest_ind(group1, group2, equal_var=False)

print(f'T-statistic: {t_stat}, P-value: {p_val}')

plt.hist(group1, alpha=0.5, label='Current Smokers', bins=15, edgecolor='black')
plt.hist(group2, alpha=0.5, label='Non-Smokers', bins=15, edgecolor='black')
plt.title('Cholesterol Levels by Smoking Status')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
